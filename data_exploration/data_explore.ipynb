# Read in the Data

```python
import pandas as pd
energy_df = pd.read_csv("../datasets/eia_data.csv")
energy_df.head()
```

```python
energy_price_df = pd.read_csv("../datasets/eia_data2.csv")
energy_price_df.head()
```

# Data Exploration

Find the sectors that are common to both datasets. Filter down to just those sectors.

```python
# find unique sectors
energy_df["sectorDescription"].unique(), energy_price_df["sectorName"].unique()
```

```python
# filter to just sectors that are in other dataset, rename appropriately
sectors = ["All Sectors", "All Industrial", "All Commercial", "Residential"]
energy_df = energy_df.loc[energy_df["sectorDescription"].isin(sectors)]
energy_df = energy_df.rename(columns={"sectorDescription": "sectorName"})
energy_df.loc[energy_df["sectorName"] == "All Sectors", "sectorName"] = "all sectors"
energy_df.loc[energy_df["sectorName"] == "All Industrial", "sectorName"] = "industrial"
energy_df.loc[energy_df["sectorName"] == "All Commercial", "sectorName"] = "commercial"
energy_df.loc[energy_df["sectorName"] == "Residential", "sectorName"] = "residential"
energy_df["sectorName"].unique()
energy_df.head()
```

```python
sectors = ["residential", "commercial", "industrial", "all sectors"]
energy_price_df = energy_price_df.loc[energy_price_df["sectorName"].isin(sectors)]
energy_price_df.head()
```

Find the fuel types that are in the data and filters down to only the ones we are interested in

```python
energy_df["fuelTypeDescription"].unique()
```

```python
#filter to relevant fuel types. We can change this later
fuel_types = ["petroleum", "natural gas", "fossil fuels", "all coal products", "biomass", "wind", "estimated total solar", "other", "other renewables", "nuclear", "hydro", "geothermal", "solar", "all fuels", "all renewables"]
energy_df = energy_df[energy_df["fuelTypeDescription"].isin(fuel_types)]
energy_df.head()
```

Checking now to see what the data looks like after that initial cleaning

```python
#row counts
energy_df.shape[0], energy_price_df.shape[0]
```

```python
#find columns with missing values
energy_df.isna().sum()
```

Start dropping some columns that aren't needed.

We aren't focused on breaking down consumption by UTO and Energy Generation, only total, so we can drop those.

We can drop the stocks, heat content, and reciepts columns because we don't care about physical quantities of material.

We can also remove the stateDescription, because the location is enough. We can also remove the sectorid because sectorName is enough.

We can also remove the physical unit columns, because BTU allows for better comparisons.


```python
energy_df.drop(columns=["stateDescription", "sectorid", "consumption-for-eg", "consumption-for-eg-units", "consumption-for-eg-btu", "consumption-for-eg-btu-units", "consumption-uto", "consumption-uto-units", "consumption-uto-btu", "consumption-uto-btu-units", "total-consumption", "total-consumption-units", "cost", "cost-units", "heat-content", "heat-content-units", "receipts", "receipts-units", "receipts-btu", "receipts-btu-units", "stocks", "stocks-units"], inplace=True)
#rename location to "stateID" to match the other data
energy_df = energy_df.rename(columns={"location": "stateid"})
energy_df.dtypes
```

Cleaning the price df

```python
energy_price_df.dtypes
```

We can remove the state description, because the stateId is enough.

Likewise, we can remove the sectorid because sectorName is enough.

The rest of the columns are important for now

```python
energy_price_df.drop(columns={"stateDescription", "sectorid"}, inplace=True)
energy_price_df.head()
```

# Bringing the Data Together

Now we can join the datasets based on sectorName, stateid, and period.

```python
final_df = pd.merge(
    energy_df,
    energy_price_df,
    on=["sectorName", "stateid", "period"],
    how="left" 
)
final_df.head()
```

Here, we can see that each row from the energy df was matched to a row in the price df

```python
final_df.shape[0]
```

Here we can see what still has missing values. 

Ash and sulfur content is to be expected since only certain fuels have that statistic.

Cost is also to be expected since solar and wind don't have a cost to generate associated. These columns might be worth removing later.

Generation is worrying because it is a crucial metric. However, the amount missing is low (82/~14000)

Consumption for electricity generation is also worrying because it is a crucial metric. The number of missing is pretty high too. 
Upon further inspection, it seems to largely be solar that is missing this metric. Our domain experts should look into why this may be the case.

# Fixing Units in Data

```python
print(final_df["total-consumption-btu-units"].value_counts())
print(final_df["ash-content-units"].value_counts())
print(final_df["generation-units"].value_counts())
print(final_df["cost-per-btu-units"].value_counts())
print(final_df["sulfur-content-units"].value_counts())
print(final_df["customers-units"].value_counts())
print(final_df["price-units"].value_counts())
print(final_df["revenue-units"].value_counts())
print(final_df["sales-units"].value_counts())
```

Upon further inspection, all the units are the same in each row. 

Therefore, I'll note the units here and drop the columns to save space.

- Total Consumption BTU  = million MMBtu
- Ash Content = Percent
- Cost Per BTU = Dollars per million Btu
- Generation = Thousand MWh
- Sulfur Content = Percent
- Customers = Number of customers
- Price = Cents per KWh
- Revenue = Million Dollars
- Sales = Million KWh

```python
final_df.drop(columns=["total-consumption-btu-units", "ash-content-units", "cost-per-btu-units", "generation-units", "sulfur-content-units", "customers-units", "price-units", "revenue-units", "sales-units"], inplace=True)
final_df.head()
```
